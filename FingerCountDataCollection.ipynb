```python
import cv2
import time
import os
import uuid
```

```python
IMAGES_PATH = "data/Images"
```

```python
labels = [str(i) for i in range(6)]
number_of_images = 35
```

```python
for label in labels:

    !mkdir {"data\\Images\\" + label}
    cap = cv2.VideoCapture(0)
    print("Collecting data for", label)
    time.sleep(2)

    for imgnum in range(number_of_images):
        print("shooting")
        time.sleep(3)
        _, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH, label, label+"{}.jpg".format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)

        if (cv2.waitKey(1) and 0Xff) == ord('q'):
            break

cap.release()
```
